Change values in price columns to numeric

```python
# Function to extract only numeric characters from a string
def extract_numbers(text):
    if isinstance(text, str):
        numeric_chars = [char for char in text if char.isdigit() or char == '.']
        return ''.join(numeric_chars)
    return text # Return original value if not a string (e.g., already a number or NaN)

# Apply the function to the 'price' column
df['dis_price'] = df['discount_price'].apply(extract_numbers)
df['act_price'] = df['actual_price'].apply(extract_numbers)

# Convert the cleaned 'price_numeric' column to a numeric type (float, int, etc.)
# Errors='coerce' will turn values that cannot be converted into NaN
df['dis_price'] = pd.to_numeric(df['dis_price'], errors='coerce')
df['act_price'] = pd.to_numeric(df['act_price'], errors='coerce')

print(df)
```

7. Relationship with Other Columns
   
Name Keywords vs. Ratings:
Do products with certain keywords in their names tend to have higher or lower ratings or no_of_ratings?
Example: Products with "pro" or "ultra" in their name: do they have higher ratings?

Name Keywords vs. Price:
Are products with specific terms (e.g., "premium", "budget") associated with higher or lower dis_price or act_price?

Name Keywords vs. Categories:
Analyze common keywords within specific main_category or sub_category to see how naming conventions differ across product types.

```python
!pip install seaborn wordcloud nltk
import nltk
nltk.download('punkt_tab')
import pandas as pd
from collections import Counter
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

# Download stopwords and tokenizer if not already
nltk.download('punkt')
nltk.download('stopwords')

# Clean column names
dfnew.columns = dfnew.columns.str.strip()

# Fill missing names if any
dfnew['name'] = dfnew['name'].fillna('')

# Define stopwords and punctuation
stop_words = set(stopwords.words('english'))
punctuations = set(string.punctuation)

# Function to clean and tokenize text
def clean_and_tokenize(text):
    text = text.lower()
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word.isalpha() and word not in stop_words and word not in punctuations]
    return tokens

# Group by main_category and process names
keywords_by_category = {}

for category, group in dfnew.groupby('main_category'):
    all_names = ' '.join(group['name'].astype(str))
    tokens = clean_and_tokenize(all_names)
    word_freq = Counter(tokens)
    keywords_by_category[category] = word_freq.most_common(10)  # Top 10 keywords

# Display keywords
for category, keywords in keywords_by_category.items():
    print(f"\nTop keywords for '{category}':")
    for word, freq in keywords:
        print(f"  {word}: {freq}")
```
