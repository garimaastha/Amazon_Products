Change values in price columns to numeric

```python
# Function to extract only numeric characters from a string
def extract_numbers(text):
    if isinstance(text, str):
        numeric_chars = [char for char in text if char.isdigit() or char == '.']
        return ''.join(numeric_chars)
    return text # Return original value if not a string (e.g., already a number or NaN)

# Apply the function to the 'price' column
df['dis_price'] = df['discount_price'].apply(extract_numbers)
df['act_price'] = df['actual_price'].apply(extract_numbers)

# Convert the cleaned 'price_numeric' column to a numeric type (float, int, etc.)
# Errors='coerce' will turn values that cannot be converted into NaN
df['dis_price'] = pd.to_numeric(df['dis_price'], errors='coerce')
df['act_price'] = pd.to_numeric(df['act_price'], errors='coerce')

print(df)
```

Check for duplicate values in the dataset (using columns name, ratings, no_of_ratings, discouted price, actual price)

```python
# Define the columns to check for duplicates
duplicate_columns = ['name', 'ratings', 'no_of_ratings', 'dis_price', 'act_price']

# Find duplicated rows based on the specified columns
duplicates = df.duplicated(subset=duplicate_columns)

# Count how many rows are duplicates
duplicate_count = duplicates.sum()

print(f'Number of duplicate rows (based on specified columns): {duplicate_count}')
```

```python
# Clean column names (optional but recommended if there are typos or spaces)
df.columns = df.columns.str.strip()

# Define the columns for uniqueness check
unique_columns = ['name', 'ratings', 'no_of_ratings', 'dis_price', 'act_price']

# Drop duplicate rows based on the 5 columns
unique_rows = df.drop_duplicates(subset=unique_columns)

# Get the count of unique rows
unique_count = len(unique_rows)

print(f'Number of unique rows (based on specified columns): {unique_count}')
```

Replace null values and string values in 'ratings' and 'no_of_ratings' with numeric zero.

```python

# Clean column names (remove leading/trailing spaces)
df.columns = df.columns.str.strip()

# Replace null (NaN) values in 'ratings' and 'no_of_ratings' with 0
df['ratings'] = pd.to_numeric(df['ratings'], errors='coerce').fillna(0)
df['no_of_ratings'] = pd.to_numeric(df['no_of_ratings'], errors='coerce').fillna(0)

# Optional: Save the cleaned data to a new CSV
# df.to_csv('cleaned_data.csv', index=False)

print("Null values in 'ratings' and 'no_of_ratings' have been replaced with 0.")

```

Replace null values and string values in price columns

```python
# Replace null dis_price with act_price
df['dis_price'] = df['dis_price'].fillna(df['act_price'])

# Replace null act_price with dis_price
df['act_price'] = df['act_price'].fillna(df['dis_price'])

# Optional: Save the result
# df.to_csv('cleaned_prices.csv', index=False)

print("Nulls in 'dis_price' and 'act_price' replaced based on each other's values.")
```
